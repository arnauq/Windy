```python
from wind import wind
import matplotlib.pyplot as plt
```

```python
wind = wind(modes = ['oldint'])
wind.thetamax = 1000
```

```python
wind.StartLines(nr =10, niter = 50000)
```

```python
plt.figure(figsize=(9,6))
for line in wind.lines:
    plt.plot(line.r_hist, line.z_hist)
plt.xlim(0,1500)
plt.ylim(0,200)
plt.xlabel("R[Rg]")
plt.ylabel("Z[Rg]")
```
